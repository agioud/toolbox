# Commande Git

| Commande Git                | Utilité                                                     |
| --------------------------- | ----------------------------------------------------------- |
| `git init`                  | Initialise un nouveau dépôt Git dans le dossier courant     |
| `git clone <url>`           | Clone un dépôt distant (GitHub, GitLab, etc.) en local      |
| `git status`                | Affiche l’état des fichiers (modifiés, ajoutés, non suivis) |
| `git add <fichier>`         | Ajoute un fichier à l’index (zone de staging)               |
| `git add .`                 | Ajoute tous les fichiers modifiés à l’index                 |
| `git commit -m "message"`   | Enregistre les changements avec un message                  |
| `git log`                   | Affiche l’historique des commits                            |
| `git log --oneline`         | Historique des commits en version courte                    |
| `git diff`                  | Montre les différences entre fichiers modifiés              |
| `git branch`                | Liste les branches existantes                               |
| `git branch <nom>`          | Crée une nouvelle branche                                   |
| `git checkout <branche>`    | Change de branche                                           |
| `git checkout -b <branche>` | Crée et change de branche en une seule commande             |
| `git switch <branche>`      | Alternative moderne à `checkout` pour changer de branche    |
| `git merge <branche>`       | Fusionne une branche dans la branche courante               |
| `git pull`                  | Récupère et fusionne les changements du dépôt distant       |
| `git push`                  | Envoie les commits locaux vers le dépôt distant             |
| `git remote -v`             | Affiche les dépôts distants configurés                      |
| `git reset <fichier>`       | Retire un fichier de l’index                                |
| `git reset --hard`          | Annule tous les changements locaux (⚠️ destructif)          |
| `git stash`                 | Met de côté les modifications non commit                    |
| `git stash pop`             | Restaure les modifications mises de côté                    |
| `git fetch`                 | Récupère les changements distants sans fusion               |
| `git rm <fichier>`          | Supprime un fichier et l’enregistre dans Git                |
| `git tag`                   | Crée ou liste des tags (versions)                           |


```python
# Ajout d'une branche local au depot distant

git switch -b nouvelle-branche   # créer la branche
git add .
git commit -m "mon commit"
git push --set-upstream origin nouvelle-branche
```
