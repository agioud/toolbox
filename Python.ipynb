````python
Les opérateurs de comparaison :
   - >  : Strictement supérieur
   - >= : Supérieur ou égal
   - <  : Strictement inférieur
   - <= : Inférieur ou égal
   - == : Egal
   - != : Différent
````

Les principaux opérateurs logiques :

and : Renvoie True si les deux opérandes sont True, sinon False. C'est comme un "ET" logique.
Exemple : (5 > 3) and (2 < 4) est True car les deux conditions sont vraies.

or : Renvoie True si au moins un des opérandes est True, sinon False. C'est comme un "OU" logique.
Exemple : (5 > 10) or (2 < 4) est True car la deuxième condition est vraie.

not : Inverse la valeur booléenne de l'opérande. Si l'opérande est True, not renvoie False, et vice versa.
Exemple : not (5 > 3) est False car 5 > 3 est True.

# LOOP

- While

La boucle **while** permet de réaliser une suite d’instructions tant qu’une certaine condition est vraie.

While= tant que

Exemple :
```
number = 0
while number < 10 :
    number += 1 # Cette notation est équivalente à number = number + 1
    print(number)
print('fini !')
```
Le code ci-dessus affiche tous les nombres entre 1 et 10, puis affiche 'fini !'

- La Boucle For...In...

**for element in iterable** signifie que la variable *element* doit prendre successivement toutes les valeurs que fournit l’*itérable*.

Ainsi, pour chacune de ces valeurs, on exécute les instructions du corps de la boucle

Exemple :
```
for number in range(0,101):
    print(number)
```
Affiche tout le nombre entier de 0 à 100


- BREAK et CONTINUE:

Dans une boucle (while ou for), l’instruction **break** interrompt la boucle instantanément : on sort de la boucle et on poursuit l’exécution du code situé après la boucle courante.

Tandis que l’instruction **continue** interrompt l’itération courante de la boucle, et passe directement à l’itération suivante, depuis le début de la boucle (pour une boucle while, on réévalue la condition, et pour une boucle for on passe à l'élément suivant)
```
for number in range(5):
 if number == 3:
  break
 print(number)
```
Affiche les nombres 0, 1 et 2. En effet, une fois que number est égal à 3 on sort de la boucle
```
for number in range(5):
 if number == 3:
  continue
 print(number)
```
Affiche tous les nombres de 0 à 4 à l'exception de 3 (on reste dans la boucle !)

# Liste

Les listes sont modifiables, ainsi il est possible de les modifier, ajouter et ou supprimer des éléments.

La fonction qui a en paramètre une liste sera en mesure de la modifier.

````python
# Definir une liste

malistevide =[]
malistevide = list()
````

**list.append(x)**

Add an item to the end of the list. Similar to a[len(a):] = [x].

**list.extend(iterable)**

Extend the list by appending all the items from the iterable. Similar to a[len(a):] = iterable.

**list.insert(i, x)**

Insère un élément à la position indiquée. Le premier argument est la position de l'élément avant lequel l'insertion doit s'effectuer, donc a.insert(0, x) insère l'élément en tête de la liste et a.insert(len(a), x) est équivalent à a.append(x).

**list.remove(x)**

Supprime de la liste le premier élément dont la valeur est égale à x. Une exception ValueError est levée s'il n'existe aucun élément avec cette valeur.

**list.pop([i])**

Remove the item at the given position in the list, and return it. If no index is specified, a.pop() removes and returns the last item in the list. It raises an IndexError if the list is empty or the index is outside the list range.

**list.clear()**

Remove all items from the list. Similar to del a[:].

**list.index(x[, start[, end]])**

Renvoie la position du premier élément de la liste dont la valeur égale x (en commençant à compter les positions à partir de zéro). Une exception ValueError est levée si aucun élément n'est trouvé.

Les arguments optionnels start et end sont interprétés de la même manière que dans la notation des tranches et sont utilisés pour limiter la recherche à une sous-séquence particulière. L'indice renvoyé est calculé relativement au début de la séquence complète et non relativement à start.

**list.count(x)**

Renvoie le nombre d'éléments ayant la valeur x dans la liste.

**list.sort(*, key=None, reverse=False)**

Ordonne les éléments dans la liste (les arguments peuvent personnaliser l'ordonnancement, voir sorted() pour leur explication).

**list.reverse()**

Inverse l'ordre des éléments dans la liste.

**list.copy()**

Return a shallow copy of the list. Similar to a[:].

````python
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
fruits.count('apple')
# renvoie 2
fruits.count('tangerine')
# renvoie 0
fruits.index('banana')
# renvoie 3
fruits.index('banana', 4)  # Find next banana starting at position 4
# renvoie 6
fruits.reverse()
# renvoie fruits['banana', 'apple', 'kiwi', 'banana', 'pear', 'apple', 'orange']
fruits.append('grape')
# renvoie fruits['banana', 'apple', 'kiwi', 'banana', 'pear', 'apple', 'orange', 'grape']
fruits.sort()
# renvoie fruits['apple', 'apple', 'banana', 'banana', 'grape', 'kiwi', 'orange', 'pear']
fruits.pop()
#renvoie 'pear'
````

# Fonction

Étapes pour créer une fonction :

1- Utiliser le mot-clé def : Cela indique que vous allez définir une fonction.

2- Nommer la fonction : Choisissez un nom qui décrit ce que fait la fonction.

3- Le nom doit être suivi de parenthèses ().

4- Ajouter des paramètres (facultatif) : Si votre fonction doit accepter des valeurs d'entrée, vous pouvez les définir entre les parenthèses.

5- Ajouter un deux-points : : Cela indique que vous allez commencer le corps de la fonction.

6- Écrire le corps de la fonction : Indentez les lignes de code qui composent la fonction. C'est ici que vous définissez ce que la fonction doit faire.

7- Retourner une valeur (facultatif) : Si vous voulez que la fonction renvoie une valeur, utilisez le mot-clé return.

-Exemple fonction

````python
def moyenne(liste):
    return sum(liste) / len(liste)

def mediane(liste):
    liste_triee = sorted(liste)
    n = len(liste_triee)
    if n % 2 == 0:
        return (liste_triee[n//2 - 1] + liste_triee[n//2]) / 2
    else:
        return liste_triee[n//2]

def ecart_type(liste):
    moy = moyenne(liste)
    variance = sum((x - moy) ** 2 for x in liste) / len(liste)
    return variance ** 0.5

def nettoyer_donnees(liste):
    return [x for x in liste if x is not None and x != ""]

def compter_occurrences(liste):
    return {x: liste.count(x) for x in set(liste)}

def normaliser(liste):
    min_val = min(liste)
    max_val = max(liste)
    return [(x - min_val) / (max_val - min_val) for x in liste]

def taux_croissance(valeur_initiale, valeur_finale):
    return (valeur_finale - valeur_initiale) / valeur_initiale * 100

def frequence_relative(liste):
    occurrences = compter_occurrences(liste)
    total = len(liste)
    return {k: v / total for k, v in occurrences.items()}

def outliers_zscore(liste, seuil=3):
    moy = moyenne(liste)
    ec = ecart_type(liste)
    return [x for x in liste if abs((x - moy) / ec) > seuil]

def trouver_mode(liste):
    occurrences = compter_occurrences(liste)
    return max(occurrences, key=occurrences.get)

def arrondir_liste(liste, decimales=2):
    return [round(x, decimales) for x in liste]
````

En Python, les blocs try et except sont utilisés pour gérer les erreurs et les exceptions. Cela permet d'écrire un code plus robuste et de prévenir les plantages du programme en cas d'erreur.

````python
# exemple simple
try:
    # Code qui pourrait provoquer une exception
    nombre = int(input("Entrez un nombre : "))
    resultat = 10 / nombre
    print("Le résultat est :", resultat)
except ValueError:
    # Gère l'erreur si l'utilisateur n'entre pas un nombre
    print("Erreur : Ce n'est pas un nombre valide.")
except ZeroDivisionError:
    # Gère l'erreur si l'utilisateur entre 0
    print("Erreur : Division par zéro n'est pas autorisée.")
````

````python
# exemple
list_integers = [1, 12, -45, 6, 27, 0, 24]
for i in list_integers:
  try : print(1/i)
  except ZeroDivisionError :
    print(f'non divisible par 0')
````

# Regex

````python
import re
````

Il existe un module re en python permettant de gérer les regex. Lorsqu'on souhaite travailler avec les regex on commence donc par importer ce module (import re) qui contient un certain nombre de fonctions.

Ci-dessous tu as quelques exemples d'utilisation de certaines fonctions du module re.

- **re.search**(motif, texte) : Renvoie un objet s'il y a une occurence du motif dans le texte

````python
import re
fournitures = "cahier stylo crayon"
re.search("stylo", fournitures)
<re.Match object; span=(7, 12), match='stylo'>     # Résultat : Objet python

# On peut ensuite utiliser une expression conditionnelle pour vérifier que tout est Ok

import re
fournitures = "cahier stylo crayon"
re.search("stylo", fournitures)
if re.search("stylo", fournitures):
  print("OK")
OK                                                 # Résultat
````

- **re.findall**(motif, texte) : Renvoie une liste contenant toutes les occurences d'un motif dans un texte

````python
import re
fournitures = "cahier stylo crayon stylo"
re.findall("stylo", fournitures)
['stylo', 'stylo']                  # Résultat renvoyé
````

- **re.split**(motif, texte) : Renvoie une liste de mots à partir du texte. Le motif sert de sepérateur pour découper le texte en mots.

````python
import re
fournitures = "cahier stylo crayon"
re.split("stylo", fournitures)
['cahier ', ' crayon']           # Résultat renvoyé
````

- **re.sub**(motif, mot_de_remplacement, texte) : Replaces one or many matches with a string

````python
import re
fournitures = "cahier stylo crayon"
re.sub("stylo", "crayon", fournitures)
'cahier crayon crayon'                 # Résultat renvoyé
````

**Les métacaractères ou caractères spéciaux ont une signification particulière lorsqu'on définit des motifs de recherche.**

. : Pour rechercher n'importe quel caractère

- Dans l'outil en ligne, recherche le motif .

? : Pour rechercher la présense d'un caractère facultatif

- Dans l'outil en ligne, recherche le motif re? Ceci va sélectionner toutes les lettres 'r' suivies ou non de la lettre 'e' (vu que la présence de 'e' ici est facultative)

^: Pour rechercher si un motif se trouve au début du texte

- Recherche dans l'outil en ligne le motif ^The

$ : Pour rechercher si un motif se trouve à la fin du texte

- Recherche dans l'outil en ligne le motif Moon.$

| : Pour rechercher si parmi deux motifs au moins un est présent dans le texte.

- Recherche dans l'outil en ligne le motif crew|team.

\ : Pour rechercher une suite de caractères ou pour échapper des caractères spéciaux

- Dans l'outil en ligne, recherche dans le texte la présence des motifs suivants \d et \&. Dans le premier cas tu vas indentifier tous les chiffres présents dans le texte, et dans le second tous les & présents dans le texte.

(*) : Pour rechercher si un motif est présent zéro ou plusieurs fois dans le texte

- Recherche dans l'outil en ligne le motif en*. Il te permet de retrouver tous les cas où on a une lettre 'e' suivie ou pas de une ou plusieurs lettres 'n'.

(+) : Pour rechercher si un motif est présent une ou plusieurs fois dans le texte

- Recherche dans l'outil en ligne le motif en+. Il te permet de retrouver tous les cas où on a une lettre 'e' suivie de 1 ou plusieurs lettres 'n'.

{} : Pour définir le nombre exact d'occurences du motif recherché dans le texte.

- Recherche dans l'outil en ligne le motif en{2}. Il te permet de retrouver tous les cas où on a une lettre 'e' suivie de deux lettres 'n'.

() : Pour rechercher un motif contenant des sous groupes de motifs.

- Recherche dans l'outil en ligne le motif ([a-z]+) (\d+). Il te permet de retrouver un groupe de une ou plusieurs lettres minuscules suivie d'un groupe de un ou plusieurs chiffres.

- Dans l'outil en ligne effectue une recherche en utilisant \Best, puis est\B

- Dans l'outil en ligne effectue une recherche en utilisant \D

- Dans l'outil en ligne effectue une recherche en utilisant \S

# subprocess.run()

````python
"""
CHEAT SHEET — subprocess.run()

Objectif :
- Lancer des commandes système depuis Python
- Chaîner des scripts (pipelines)
- Gérer succès / échec / sorties

Contexte typique :
- Pipeline bioinfo
- Wrapper CLI
- Orchestration de scripts Python
"""

import subprocess

# ============================================================
# 1) USAGE MINIMAL
# ============================================================

# Lance une commande comme dans un terminal :
# $ python script.py P115
subprocess.run(["python", "script.py", "P115"])


# ============================================================
# 2) RÉCUPÉRER LE RÉSULTAT
# ============================================================

result = subprocess.run(["python", "script.py", "P115"])

# Code de sortie (0 = succès, != 0 = erreur)
print(result.returncode)


# ============================================================
# 3) TESTER SUCCÈS / ÉCHEC
# ============================================================

if result.returncode == 0:
    print("✅ Succès")
else:
    print("❌ Erreur")


# ============================================================
# 4) CAPTURER stdout / stderr
# ============================================================

result = subprocess.run(
    ["python", "script.py", "P115"],
    capture_output=True,   # capture stdout + stderr
    text=True              # renvoie str au lieu de bytes
)

print("STDOUT:")
print(result.stdout)

print("STDERR:")
print(result.stderr)


# ============================================================
# 5) STOPPER LE SCRIPT EN CAS D’ERREUR
# ============================================================

try:
    subprocess.run(
        ["python", "script.py", "P115"],
        check=True           # lève une exception si returncode != 0
    )
except subprocess.CalledProcessError as e:
    print("❌ Le script a échoué")
    print("Code retour :", e.returncode)


# ============================================================
# 6) TIMEOUT (ANTI-BLOCAGE)
# ============================================================

try:
    subprocess.run(
        ["python", "script.py"],
        timeout=60           # secondes max
    )
except subprocess.TimeoutExpired:
    print("⏱️ Timeout dépassé")


# ============================================================
# 7) CHANGER LE DOSSIER D’EXÉCUTION
# ============================================================

subprocess.run(
    ["python", "script.py"],
    cwd="A:/projet/sv_filter"   # équivalent à un `cd` avant exécution
)


# ============================================================
# 8) VARIABLES D’ENVIRONNEMENT
# ============================================================

subprocess.run(
    ["python", "script.py"],
    env={"MODE": "DEBUG"}       # remplace l’environnement
)


# ============================================================
# 9) MASQUER LA SORTIE (MODE SILENCIEUX)
# ============================================================

subprocess.run(
    ["python", "script.py"],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL
)


# ============================================================
# 10) FONCTION PIPELINE PROPRE (TON CAS)
# ============================================================

def run(script, patient):
    """
    Lance un script Python pour un patient
    Retourne True si succès, False sinon
    """
    print(f"🚀 {script} — {patient}")

    result = subprocess.run(
        ["python", script, patient]
    )

    return result.returncode == 0
````
