# Les Bases

Scala est un langage :

✔ Orienté Objet (comme Java)

✔ Fonctionnel (comme Python, JavaScript, Haskell)

✔ Très utilisé dans les data pipelines (Apache Spark)

Il fonctionne sur la JVM (machine virtuelle Java).

## Imprimer du texte
```scala
println("Hello Scala!")
```

## Variables
```scala
val x = 10     // immuable
var y = 5      // mutable
```

## Fonctions
```scala
def add(a: Int, b: Int): Int =
  a + b
```

## Conditions
```scala
val age = 16
if age >= 18 then println("Majeur") else println("Mineur")
```

## Listes
```scala
val nums = List(1, 2, 3)
val doubled = nums.map(_ * 2)
val even = nums.filter(_ % 2 == 0)
```

## Pattern Matching
```scala
val note = 15
val mention = note match
  case 20 => "Parfait"
  case n if n >= 15 => "Bien"
  case n if n >= 10 => "Moyen"
  case _ => "Insuffisant"
```
