```python
üìÇ Navigation
pwd              ‚Üí dossier actuel
ls / ls -l / ls -a ‚Üí liste fichiers
cd dossier       ‚Üí entrer dossier
cd ..            ‚Üí niveau au-dessus
cd ~             ‚Üí home
cd -             ‚Üí dossier pr√©c√©dent

üìÅ Fichiers & Dossiers
touch file            ‚Üí cr√©er fichier
code file             ‚Üí cr√©er fichier vs
mkdir dir             ‚Üí cr√©er dossier
mkdir -p a/b/c        ‚Üí cr√©er arborescence

cp file copy          ‚Üí copier fichier
cp -r dir copy        ‚Üí copier dossier

mv file dir/          ‚Üí d√©placer
mv old new            ‚Üí renommer

rm file               ‚Üí supprimer fichier
rm -r dir             ‚Üí supprimer dossier
rm -rf dir            ‚Üí suppression forc√©e ‚ö†Ô∏è

üìñ Lecture fichiers
cat file        ‚Üí afficher contenu
less file       ‚Üí lecture scrollable
head file       ‚Üí d√©but fichier
tail file       ‚Üí fin fichier
tail -f log     ‚Üí suivi temps r√©el

üîé Recherche
find . -name "file"       ‚Üí chercher fichier
grep "text" file          ‚Üí chercher texte
grep -r "text" folder     ‚Üí recherche r√©cursive

‚öôÔ∏è Permissions
chmod +x script.sh    ‚Üí rendre ex√©cutable
chmod 755 file        ‚Üí modifier droits
chown user file       ‚Üí changer propri√©taire

üì¶ Processus
ps           ‚Üí liste processus
top / htop   ‚Üí monitoring live
kill PID     ‚Üí stop processus
kill -9 PID  ‚Üí stop forc√©

üåê R√©seau
ping site.com      ‚Üí tester connexion
curl URL           ‚Üí requ√™te HTTP
wget URL           ‚Üí t√©l√©charger fichier

üíæ Syst√®me & stockage
df -h          ‚Üí espace disque
du -sh dir     ‚Üí taille dossier
free -h        ‚Üí RAM
uname -a       ‚Üí infos syst√®me

üîÑ Redirections & Pipes ‚≠ê
cmd > file        ‚Üí sortie vers fichier (√©crase)
cmd >> file       ‚Üí ajoute sortie
cmd1 | cmd2       ‚Üí pipeline


Exemple :

ls -l | grep ".txt"

üß† Variables
echo $VAR
export VAR=value
env

‚å®Ô∏è Raccourcis Terminal
Ctrl + C    ‚Üí stop
Ctrl + D    ‚Üí quitter
Ctrl + L    ‚Üí clear
Ctrl + R    ‚Üí recherche historique
Tab         ‚Üí autocompl√©tion

üî• Commandes tr√®s utiles
sudo cmd     ‚Üí admin
man cmd      ‚Üí documentation
which cmd    ‚Üí chemin commande
history      ‚Üí historique
alias ll="ls -l"
!!           ‚Üí derni√®re commande

‚≠ê Rappels importants
‚ö†Ô∏è rm -rf peut supprimer TOUT
üí° utiliser TAB souvent
üí° Ctrl + R = √©norme gain de temps
```

# Commande Git

| Commande Git                | Utilit√©                                                     |
| --------------------------- | ----------------------------------------------------------- |
| `git init`                  | Initialise un nouveau d√©p√¥t Git dans le dossier courant     |
| `git clone <url>`           | Clone un d√©p√¥t distant (GitHub, GitLab, etc.) en local      |
| `git status`                | Affiche l‚Äô√©tat des fichiers (modifi√©s, ajout√©s, non suivis) |
| `git add <fichier>`         | Ajoute un fichier √† l‚Äôindex (zone de staging)               |
| `git add .`                 | Ajoute tous les fichiers modifi√©s √† l‚Äôindex                 |
| `git commit -m "message"`   | Enregistre les changements avec un message                  |
| `git log`                   | Affiche l‚Äôhistorique des commits                            |
| `git log --oneline`         | Historique des commits en version courte                    |
| `git diff`                  | Montre les diff√©rences entre fichiers modifi√©s              |
| `git branch`                | Liste les branches existantes                               |
| `git branch <nom>`          | Cr√©e une nouvelle branche                                   |
| `git checkout <branche>`    | Change de branche                                           |
| `git checkout -b <branche>` | Cr√©e et change de branche en une seule commande             |
| `git switch <branche>`      | Alternative moderne √† `checkout` pour changer de branche    |
| `git merge <branche>`       | Fusionne une branche dans la branche courante               |
| `git pull`                  | R√©cup√®re et fusionne les changements du d√©p√¥t distant       |
| `git push`                  | Envoie les commits locaux vers le d√©p√¥t distant             |
| `git remote -v`             | Affiche les d√©p√¥ts distants configur√©s                      |
| `git reset <fichier>`       | Retire un fichier de l‚Äôindex                                |
| `git reset --hard`          | Annule tous les changements locaux (‚ö†Ô∏è destructif)          |
| `git stash`                 | Met de c√¥t√© les modifications non commit                    |
| `git stash pop`             | Restaure les modifications mises de c√¥t√©                    |
| `git fetch`                 | R√©cup√®re les changements distants sans fusion               |
| `git rm <fichier>`          | Supprime un fichier et l‚Äôenregistre dans Git                |
| `git tag`                   | Cr√©e ou liste des tags (versions)                           |
